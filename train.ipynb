```python
import pandas as pd
```

```python
df = pd.read_csv('apple_stock_real_price.csv', delimiter='|')
df
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop(columns=['Unnamed: 0', 'tick_volume', 'spread', 'real_volume'], inplace=True)
df
```

```python
for lag in range(1, 4):
    df[f"close_lag{lag}"] = df["close"].shift(lag)
```

```python
df
```

```python
df = df.dropna()
```

```python
df["time"] = pd.to_datetime(df["time"])
df["year"] = df["time"].dt.year
df["month"] = df["time"].dt.month
df["day"] = df["time"].dt.day
df["dayofweek"] = df["time"].dt.dayofweek
df
```

```python
X = df.drop(columns=['close', 'time'])
y = df['close']
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
X_scaled
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X = X_scaled, y = y)
y_train_pred = lr.predict(X = X_scaled)

y
```

```python
y_train_pred = pd.Series(y_train_pred)
y_train_pred
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(f"MSE error in training data : {mean_squared_error(y_true = y, y_pred = y_train_pred)}")
print(f"MAE error in training data : {mean_absolute_error(y_true = y, y_pred = y_train_pred)}")
```
